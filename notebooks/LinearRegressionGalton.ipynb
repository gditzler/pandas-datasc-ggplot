# About

This notebook highlights linear regression on the `Galton.csv` data, which contains a list of individual's heights. It contains the heights of parents and their children. There are 928 observations on the aforementioned variables. The midparent's height is an average of the fathers height and 1.08 times the mother's. In the data there are 205 different parents and 928 children.  

# Implementing Linear Regression Using Pandas

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm 
from ggplot import * 
```

Load the data from the `../data/` directory using Pandas. Then print out the header of the data and plot the parents' height versus the childs height. We are going to use a linear model to predict the height of the child given the height of the parent. 

```python
df = pd.read_csv("../data/Galton.csv")
print df.head()
ggplot(aes(x="Midparent", y="Child"), data=df) + geom_point(colour="blue") 
```

We can also use the builtin functionality of the `scatter_matrix` for Pandas to visualize the data being plotted against each other as well as the probabilty density functions of each of the variables in the data. 

```python
pd.scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal='kde') 
```

```python
X = df["Midparent"]
Y = df["Child"]
model = sm.OLS(Y, X)
results = model.fit() 
results.summary()
```

```python
df2 = pd.DataFrame({"Prediction": results.fittedvalues}) 
df3 = pd.concat([df, df2],axis=1)  
df3.head()
ggplot(aes(x="Midparent", y="Child"), data=df3) + geom_point(colour="red") + stat_smooth(method='lm', color='blue')  
```
