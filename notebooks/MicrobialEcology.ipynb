# Simple Data Analysis in Microbial Ecology

## Importing the Required Python Modules

```python
import pandas as pd
import numpy as np
import sys 
from scipy.spatial.distance import * 
from sklearn import manifold 
import matplotlib.pylab as plt
sys.path.append("../src/")
from load_biom import load_biom 
```

## Loading the Data and Preparing for Pandas

this is ... $x$

```python
meta_data = pd.read_csv("../data/caporaso-gut.txt", sep="\t")
data, sample_ids, otus = load_biom("../data/caporaso-gut.biom")

# we need to convert the raw biom data into a dataframe
df = {}
for n,id in enumerate(sample_ids): 
    df[id] = pd.Series(data[n], index = otus)
df = pd.DataFrame(df)
```

## Multi-Dimensional Scaling 

this is another

```python
X = np.array([df[x].tolist() for x in df.keys()])
dist_mat = squareform(pdist(X, "cosine")) 
mds = manifold.MDS(n_components=2, dissimilarity="precomputed", random_state=3, n_jobs = 2)
results = mds.fit(dist_mat) 
coords = results.embedding_ 
```

## Visualizing the Results

this is 

```python
dimred = pd.DataFrame({"PC1": pd.Series(coords.transpose()[0].tolist(), index=sample_ids), \
     "PC2": pd.Series(coords.transpose()[1].tolist(), index=sample_ids)})
plt.plot(dimred["PC1"], dimred["PC2"], marker="o", lw=0)
```

Separating out the males and females

```python
idx_male = np.where("male" == np.array(meta_data["SEX"].tolist()))[0]
idx_female = np.where("female" == np.array(meta_data["SEX"].tolist()))[0]

ids_female = [ sample_ids[i] for i in idx_female]
ids_male = [ sample_ids[i] for i in idx_male]

plt.plot(dimred["PC1"][ids_male], dimred["PC2"][ids_male], marker="o", c="b", lw=0)
plt.plot(dimred["PC1"][ids_female], dimred["PC2"][ids_female], marker="o", c="r", lw=0)
```
